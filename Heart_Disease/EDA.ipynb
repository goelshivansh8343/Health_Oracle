```python
x=df.columns
columns=x.drop("target")
plt.figure(figsize=(15,10))
for i in range(len(columns)):
    plt.subplot(4,4,i+1)
    sns.histplot(data=df,x=columns[i])


plt.show()
```

```python
plt.figure(figsize=(15,10))
for i in range(len(columns)):
    plt.subplot(4,4,i+1)
    sns.boxplot(data=df,x=columns[i])

plt.show()
```
